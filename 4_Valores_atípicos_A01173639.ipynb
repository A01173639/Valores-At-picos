**PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=cuantitativas_2020
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data3= cuantitativas_2020[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data3.copy()
data_clean_2020_MDE=data_clean.fillna(round(data3.mean(),1))
data_clean_2020_MDE
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean_2020_MDE.isnull().sum()
valores_nulos
```

**PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=cuantitativas_2020

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= cuantitativas_2020[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data5_iqr=data3_iqr.copy()
data5_iqr=data5_iqr.fillna(round(data3_iqr.mean(),1))
data5_iqr
```

```python
# Unimos  los datos cualitativos con el dataframe

c=cual_2020
data_clean_2020_IQR= pd.concat([c, data5_iqr], axis=1)
data_clean_2020_IQR
```
